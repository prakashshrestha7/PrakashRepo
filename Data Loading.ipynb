```python
#Loading Libraries
import pandas as pd
import matplotlib.pyplot as plt

#Histogram Plot function
def distribution(plot_data, column_name):
    plot_data[column_name] = plot_data[column_name].astype('category')
    category_codes = plot_data[column_name].cat.codes
    bins = plot_data[column_name].cat.categories
    plt.hist(plot_data[column_name].astype(str), bins = len(bins))
    plt.xticks(rotation = 90)
    plt.title('Channel type distribution of top 1000 youtube channel records')
    plt.xlabel('Channel Type')
    plt.show()

#Data Slicing Fucntion    
def data_slicer(data_frame, first_row, last_row, first_column, last_column):
    filtered_data = data_frame.iloc[first_row:last_row, first_column:last_column]
    distribution(filtered_data, 'channeltype') 
    return filtered_data

#Data export to CSV function
def csv_writer(data, output_filename):
    data.to_csv(output_filename, index = False)
    print("Data successfully written to file {}".format(output_filename))
    
df = pd.read_csv("./youtube_dataset.csv", encoding = 'windows-1252') #To handle euro symbol in Earnings columns
df.head()    
```

```python
top_1000 = data_slicer(df, 0, 1000, 0, len(df.axes[1]))
```

```python
csv_writer(top_1000, 'top_1000_records.csv')
```
